## 1) Regressão Linear - intuição

Hoje estudaremos modelos de previsão de séries temporais. Os modelos que estudaremos serão utilizados para resolver um tipo de problema que é classificado como um **problema de regressão**.

Problemas de regressão são aqueles em que o que desejamos modelar/prever **é um valor numérico contínuo**, que no nosso caso será o próprio valor da série temporal (em instantes futuros).

O tipo mais simples de modelo de regressão é a **regressão linear**, que tem como objetivo **determinar qual é a reta que melhor se aproxima aos dados**.

Como o próprio nome diz, o modelo de Regressão Linear será **uma reta (polinômio linear)**, que melhor se ajusta aos seus dados!

> O modelo de **Regressão Linear Simples** será uma linha reta que relaciona **y** (que é o que queremos prever, o autput, comumente chamado de **target**) com **X** (os atributos utilizados para determinar o target, comumente chamados de **preditores** ou **features**). 

Se utilizarmos **apenas um atributo**, temos uma **Regressão Linear Simples**, e nosso modelo é:

$$ y = b_0 + b_1 X $$

Neste caso, o modelo tem dois coeficientes a serem determinados: $b_0$ (intercepto ou coeficiente linear) e $b_1$ (coeficiente angular). 

Se, por outro lado, tivermos mais de um preditor, o modelo passa a ser a **Regressão Linear Múltipla**:

$$ y = b_0 + b_1 X_1 + b_2 X_2 + \cdots + b_n X_n $$

Neste caso, além de $b_0$ e $b_1$, temos também outros coeficientes, um pra cada uma das $n$ features que escolhermos!

O algoritmo do estimador é utilizado justamente para encontrarmos os coeficientes $\{b_i\}$ **que melhor se ajustam aos dados!**

Para fazer isso, pode-se utilizar o método dos **mínimos quadrados (least squares)** ou então o **gradiente descendente**.

<img src="https://qph.fs.quoracdn.net/main-qimg-3b0d7655ac76edf1241f97015ee755b4" width=450>

Os modelos preditivos de séries temporais que veremos a seguir, são modelos de regressão -- mas cujos preditores são os **valores passados da própria série**. É por isso que estes modelos são chamados de **autorregressivos**, pois o target e as features se referem à mesma quantidade: a própria série temporal! Com a importante diferença de que, naturalmente, o target objetiva valores **futuros** da série.

Vamos agora conhecer estes modelos!

## 2) Autocorrelação

Um dos conceitos mais importantes que permeará nossos modelos autorregressivos será o de **autocorrelação**, que nada mais é que a **correlação da série temporal com uma cópia dela mesma**, mas de uma forma **defasada** (ou seja, tomando cópias intervalados por períodos denominados **lags**):

<img src="https://i1.wp.com/www.business-science.io/assets/lag-explanation.PNG?w=584" width=400>

A depender do sinal da autocorreção, temos comportamentos diferentes:

- autocorrelação positiva: **momentum** ou **trend following**
- autocorrelação negativa: **mean reversion**

<img src="https://46gyn61z4i0t1u1pnq2bbk2e-wpengine.netdna-ssl.com/wp-content/uploads/2018/04/Positive-and-negative-autocorrelation.png" width=450>

Para visualizar a autocorrelação, é muito comum que utilizemos os plots **ACF** e **PACF**:

____

### 2.1) Função de Autocorrelação - Auto Correlation Function (ACF)

A função de autocorrelação, ACF, é calculada considerando **todos os instantes de tempo intermediários dentro do lag a partir do primeiro instante de tempo**.

É muito comum visualizá-la com auxílio da biblioteca statsmodels:

```python
# importando o método
from statsmodels.graphics.tsaplots import plot_acf
```

```python
precos[:35].plot()
```

```python
plot_acf(precos)

plt.show()
```

### 2.2) Função de Autocorrelação Parcial - Partial Auto Correlation Function (PACF)

A função de autocorrelação parcial, PACF, é calculada considerando **apenas os instantes de tempo do lag, sem considerar os instantes intermediários**.

Também há um método do statsmodels para visualizarmos a PACF:

```python
# importando o método
from statsmodels.graphics.tsaplots import plot_pacf
```

```python
precos[:35].plot()
```

### 2.3) White Noise

Um importante componente dos modelos que criaremos será o **white noise** (ruído branco), que será responsável por capturar as variações aleatórias na série, que não podem ser estatisticamente modeladas.

O white noise é caracterizado por ter:

- média constante;
- variância constante;
- autocorrelação zero em todas lags.

Um exemplo de white noise:

<img src="https://3qeqpr26caki16dnhd19sv6by6v-wpengine.netdna-ssl.com/wp-content/uploads/2017/01/White-Noise-Series-Line-Plot.png" width=500>


Uma forma simples de construir white noise é a partir deums distribuição normal:

```python
np.random.seed(42)
noise = pd.Series(np.random.normal(loc=0, scale=1, size=500))
```

```python
sns.histplot(noise)
```

### 2.4) Random Walk

Além do white noise, um tipo de série temporal de grande interesse são os **random walks** (**passeios aleatórios**).

Diferentemente do white noise, que consiste em **valores sucessivos aleatórios e descorrelacionados**, os passeios aleatórios consistem em sequências tais que **valores seguintes são construídos a partir de valores anteriores**. Por este motivo, passeios aleatórios são **modeláveis**, formando a base dos modelos que estudaremos.

O fato de haver dependência entre valores sucessivos de um passeio aleatório faz com que haja correlação entre os diferentes valores, o que faz com que as variações sejam mais suaves que os "pulos" descontínuos do white noise.

A forma mais simples de modelar um passeio aleatório é tomando o preço de hoje como sendo o preço de ontem + ruído, além de um termo adicional independente chamado de **drift**, isto é,

$$P_t = \mu + P_{t-1} + \epsilon_t$$

Ou seja, a mudança de preço é ruído branco com média $\mu$:

$$P_t - P_{t-1} = \mu + \epsilon_t$$

Para saber mais sobre random walks, [clique aqui](https://machinelearningmastery.com/gentle-introduction-random-walk-times-series-forecasting-python/)

### 2.5) Estacionariedade

Uma série temporal é dita **estacionária** quando ela se desenvolve no tempo aleatoriamente ao redor de uma **média constante**, e com **variância constante**, refletindo um equilíbrio estável:

<img src="https://miro.medium.com/max/2580/1*tkx0_wwQ2JT7pSlTeg4yzg.png" width=550><br>

<img src="https://miro.medium.com/max/1147/1*xdblkZyg6YmmReAkZHUksw.png" width=650>


> -**Random walks** são exemplos de séries **não estacionárias**:

<img src="https://3qeqpr26caki16dnhd19sv6by6v-wpengine.netdna-ssl.com/wp-content/uploads/2017/01/Random-Walk-Line-Plot.png" width=450>

> **Séries sazonais** também não são estacionárias:

<img src="https://anomaly.io/wp-content/uploads/2015/12/multiplicative-model.png" width=450>

> - **White noise** é estacionário:

<img src="https://www.researchgate.net/publication/279249485/figure/fig10/AS:272660147994627@1442018653078/An-example-of-a-stationary-time-series-specifically-a-series-of-uncorrelated-white.png" width=500>

Para avaliarmos se uma série temporal é ou não estacionária, fazemos o [teste de Dickey-Fuller](https://en.wikipedia.org/wiki/Dickey%E2%80%93Fuller_test).

```python
from statsmodels.tsa.stattools import adfuller

adfuller(precos)
```

```python
# p-value menor que 5% e rejeitamos a hipótese nula que a 
# série é um random walk com 95% de confiança

p = adfuller(precos)[1]

print("\np-value:", p)

# interpretação do teste
# nível de confiança de 5%
alpha = 0.05

if p > alpha:
    print("\nFalha em rejeitar H_0: amostra parece ser não-estacionaria")
else:
    print("\nRejeição da H_0: amostra é estacionária")
```

Para **tornarmos uma série não-estacionária estacionária**, há um procedimento muito simples que modemos tentar fazer:
tomar as variações entre diferentes instantes de tempo, o que fazemos com o método `diff()`, que é bem similar ao `pct_change()`, exceto pelo fato de que não há a divisão pelo instante anterior:

> `diff()`: $P_t - P_{t-1}$

> `pct_change()`: $\frac{P_t - P_{t-1}}{P_{t-1}}$

```python
difs = precos.diff().dropna()

difs
```

```python
plt.figure(figsize=(14, 6))

sns.lineplot(x=difs.index, y=difs, label="ST de diferenças")

plt.legend()
plt.show()
```

Ou podemos usar diretamente a própria série de retornos:


```python
retornos = precos.pct_change().dropna()

plt.figure(figsize=(14, 6))

sns.lineplot(x=retornos.index, y=retornos, label="ST de retornos")

plt.legend()
plt.show()
```

Agora que conhecemos o importanto processo de estacionaridade, vamos introduzir dois procedimentos que são utilizados **para modelar processos estacionários**.

```python
p = adfuller(difs)[1]

print("\np-value:", p)

# interpretação do teste
# nível de confiança de 5%
alpha = 0.05

if p > alpha:
    print("\nFalha em rejeitar H_0: amostra parece ser não-estacionaria")
else:
    print("\nRejeição da H_0: amostra é estacionária")
```

## 3) Modelos AR

O primeiro modelo que vamos tratar é o **Modelo AutoRegressivo** (daí o **AR**). 

O objetivo deste (e dos demais modelos que veremos) é **capturar o padrão por trás da série e prever o comportamento futuro a partir do passado**.

Comparação com random walk:

$$R_t = \mu + R_{t-1} + \epsilon_t$$

Por sua vez, este modelo estabelece uma relação muito precisa entre diferentes instantes de tempo, expressa a seguir como o **modelo autorregressivo de ordem $p$ - $\text{AR}(p)$:**


$$R_t = \mu + \sum_{i=1}^p \phi_i R_{t-i} + \epsilon_t \ , $$

Ou seja,

$$R_t = \mu + \phi_1 R_{t-1} + \phi_2 R_{t-2} + \cdots + \phi_p R_{t-p} + \epsilon_t \ , $$

Onde:

- $R_t$ é o instante atual;
- $p$ é o número de lags considerado no modelo (também chamado de **ordem** do modelo);
- $R_{t-p}$ é o instante de tempo passados p lags;
- $\phi_p$ é o respectivo coeficiente da PACF.

Um bom modelo será um que utiliza um número $p$ de lags da PACF fora do cone de confiança (ou seja, estatisticamente significantes).

Considere o modelo $\text{AR}(1)$ (ou seja, que considera apenas um lag $p=1$). Temos $R_t = \mu + \phi_1 R_{t-1}
+ \epsilon_t$, de modo que:

> - $\phi_1 = 1$: random walk;
> - $\phi_1 = 0$: white noise;
> - estacionaridade: $-1 < \phi_1 < 1$;
> - $\phi_1$ negativo: mean reversion;
> - $\phi_1$ positivo: momentum.


A primeira coisa antes de construirmos nosso modelo, será definir qual será nosso **conjunto de teste**:

```python
num_test = 30

retornos_test = retornos.iloc[-num_test:]
```

```python
retornos_test
```

Para treinarmos um modelo $\text{AR}(p)$ para nossa série temporal, vamos utilizar o submódulo `tatsmodels.tsa.arima_model` da [statsmodels](https://www.statsmodels.org/stable/tsa.html)


```python
from statsmodels.tsa.arima_model import ARMA
```

**Identificando a ordem $p$ do modelo $\text{AR}(p)$**

O primeiro passo para a construção do modelo é a determinação da **ordem $p$**, ou seja, o número de lags em termos dos quais vamos modelar o target.

Para fazer isso, a principal abordagem é avaliar o gráfico da **PACF**, e escolher lags que sejam estatisticamente dignificantes! 

```python
plot_pacf(retornos)
plt.show()
```

ARMA(p, 0) = AR(p)

ARMA(0, q) = MA(q)

```python
modelo = ARMA(retornos, order=(1, 0)).fit()
```

Após fitarmos o modelo, podemos dar uma olhada em informações com o método `summary()`:

```python
print(modelo.summary())
```

Agora, podemos usar o modelo para fazer predições **nos dados de teste**:

```python
retornos_test.head()
```

```python
preds = modelo.predict(start=retornos_test.index[0], end=retornos_test.index[-1]).head()
```

É importante olharmos pros resíduos nos dados de teste, para garantirmos que os resíduos não estão conentrados em positivos ou negativos:

```python
preds = modelo.predict(start=retornos_test.index[0], end=retornos_test.index[-1])

# residuos do modelo
residuos = retornos_test - preds

plt.figure(figsize=(12,6))
plt.plot(residuos)
plt.ylabel('$\epsilon$')
plt.axhline(0, color='r', linestyle=':')
plt.axhline(residuos.mean(), color='green', linestyle=':')
```

Vamos visualizar as predições juntamente com os valores reais:

```python
plt.figure(figsize=(14, 6))
plt.title(simbolo + " - série de retornos")

sns.lineplot(x=retornos_test.index, y=retornos_test, label="retornos originais")

preds = modelo.predict(start=retornos_test.index[0], 
                       end=retornos_test.index[-1] + dt.timedelta(days=1))
sns.lineplot(x=preds.index, y=preds, label="retornos preditos")

plt.legend()
plt.show()
```

```python
modelo.plot_predict(start=retornos_test.index[0], 
                       end=retornos_test.index[-1] + dt.timedelta(days=10))

plt.show()
```

O modelo parece que não funcionou nada bem... De fato, modelos AR funcionariam super bem caso houvesse **clara sazonalidade** na série. Como este não é o caso de nossa série, era esperado que este modelo não performasse tão bem.

## 4) Modelos MA

Vamos tratar agora do **Modelo de Média Móvel** (**Moving Average**, daí o **MA**). 

Diferente dos modelos AR, os modelos MA estabelecem uma relação não entre os valores de diferentes instantes de tempo, mas sim entre **os ruídos de diferentes instantes de tempo**. Esta rellação é xpressa a seguir como o **modelo de média móvel de ordem $q$ - $\text{MA}(q)$:**

$$R_t = \mu + \sum_{i=1}^q \theta_i \epsilon_{t-i} + \epsilon_t \ , $$

Ou seja,

$$R_t = \mu + \theta_1 \epsilon_{t-1} + \theta_2\epsilon_{t-2} + \cdots + \theta_q \epsilon_{t-q} + \epsilon_t \ , $$

Onde:

- $R_t$ é o instante atual;
- $q$ é o número de lags considerado no modelo (também chamado de **ordem** do modelo);
- $\epsilon_{t-q}$ é o ruído entre o instante atual e o instante após q lags;
- $\theta_q$ é o respectivo coeficiente da ACF.

Um bom modelo será um que utiliza um número $q$ de lags da ACF fora do cone de confiança (ou seja, estatisticamente significantes).


Vamos construir agora um modelo MA para nossa série temporal!

A primeira coisa é avaliarmos a ordem adequada com base na ACF:

```python
plot_acf(retornos)
plt.show()
```

```python
modelo = ARMA(retornos, order=(0, 1)).fit()
```

```python
print(modelo.summary())
```

```python
modelo.predict(start=retornos_test.index[0], end=retornos_test.index[-1])
```

```python
preds = modelo.predict(start=retornos_test.index[0], end=retornos_test.index[-1])

# residuos do modelo
residuos = retornos_test - preds

plt.figure(figsize=(12,6))
plt.plot(residuos)
plt.ylabel('$\epsilon$')
plt.axhline(0, color='r', linestyle=':')
```

```python
plt.figure(figsize=(14, 6))
plt.title(simbolo + " - série de retornos")

sns.lineplot(x=retornos_test.index, y=retornos_test, label="retornos originais")
sns.lineplot(x=preds.index, y=preds, label="retornos preditos")

plt.legend()
plt.show()
```

## 5) Modelos ARMA

Os modelos ARMA, como o nome indica, são modelos compostos por ambas as componentes AR e MA. O modelo $\text{ARMA}(p, q)$ é dado por:

$$R_t = \mu + \sum_{i=1}^p \phi_i R_{t-i} + \sum_{i=1}^q \theta_i \epsilon_{t-i} + \epsilon_t \ . $$

Uma pergunta natural é: em que caso usar qual modelo?

### Information Criteria

Uma pergunta natural é: como escolher a ordem dos modelos?

Uma abordagem é utilizar os **critérios de informação**, que são métricas cujo objetivo é estimar a **qualidade de um modelo**, mas de forma inversa, isto é, **quanto menores os valores das métricas dos critérios, melhor é o modelo**.

Os principais critérios utilizados são o **AIC - Akaike Information Criterion** e o **BIC - Bayesian Information Criterion**.

O que podemos fazer é fitar modelos de diferentes ordens, e escolher como modelo final aquele com menores IC!

Para saber mais sobre seleção de modelos, [clique aqui](https://machinelearningmastery.com/probabilistic-model-selection-measures/), e sobre os AIC/BIC, [clique aqui.](https://coolstatsblog.com/2013/08/14/using-aic-to-test-arima-models-2/)

Obs.: este é apenas **um dos critérios** de seleção de modelos. Nunca se baseie apenas neles!

```python
ls_ar, ls_ma, ls_aic, ls_bic = [], [], [], []

for ar in range(0, 3):
    for ma in range(0, 3):
        
        try:
            mod = ARMA(retornos, order=(ar, ma)).fit()

            ls_ar.append(ar)
            ls_ma.append(ma)
            ls_aic.append(mod.aic)
            ls_bic.append(mod.bic)
            
        except:
            
            pass
```

```python
df_ic = pd.DataFrame({'AR': ls_ar, 'MA': ls_ma, 'AIC': ls_aic, 'BIC': ls_bic})
```

```python
df_ic.sort_values(['AIC', 'BIC'])
```

## 6) Modelos ARIMA

Por fim, chegamos ao modelo mais complexo que aprenderemos, o **AutoRegressive Integrated Moving Average**.

Estes modelos se diferenciam do ARMA ao incorporar o procedimento de **diferenciação** da série temporal **original e não estacionária**. Este procedimento faz com que a série se torne estacionária para que o forecasting seja feito. Então, como passo final, a diferenciação é revertida com a **integração**, o que faz com que as predições sejam feitas em termos dos valores originais.

Assim, poderemos com o ARIMA utilizar diretamente as séries de preços!

Os modelos $\text{ARIMA}(p, d, q)$ têm as seguintes componentes:

- p: número de defasagens autoregressivas;
- d: ordem da difereenciação das defasagens;
- q: número de defasagens das médias móveis.

Vamos à construção do modelo!

```python
from statsmodels.tsa.arima_model import ARIMA
```

```python
num_test = 50

precos_test = precos.iloc[-num_test:]
```

```python
modelo = ARIMA(precos, order=(1, 1, 1)).fit()
```

```python
print(modelo.summary())
```

É importante olharmos pros resíduos nos dados de teste, para garantirmos que os resíduos não estão conentrados em positivos ou negativos:

```python
preds = modelo.predict(start=precos_test.index[0], end=precos_test.index[-1], typ="levels")

# residuos do modelo
residuos = precos_test - preds

plt.figure(figsize=(12,6))
plt.plot(residuos)
plt.ylabel('$\epsilon$')
plt.axhline(0, color='r', linestyle=':')
```

Vamos visualizar as predições juntamente com os valores reais:

```python
plt.figure(figsize=(14, 6))
plt.title(simbolo + " - série de retornos")

sns.lineplot(x=precos_test.index, y=precos_test, label="retornos originais")
sns.lineplot(x=preds.index, y=preds, label="retornos preditos")

plt.legend()
plt.show()
```

```python
preds = modelo.predict(start=precos_test.index[0], 
                       end=precos_test.index[-1] + dt.timedelta(days=2), typ="levels")

plt.figure(figsize=(14, 6))
plt.title(simbolo + " - série de retornos")

sns.lineplot(x=precos_test.index, y=precos_test, label="retornos originais")
sns.lineplot(x=preds.index, y=preds, label="retornos preditos")

plt.legend()
plt.show()
```

```python
modelo.plot_predict(start=precos_test.index[0], 
                    end=precos_test.index[-1] + dt.timedelta(days=2))
plt.show()
```

Vamos ver um contexto onde o ARIMA funcionará super bem!

```python
df_flights = sns.load_dataset('flights')
df_flights.index = pd.date_range('1949-01', periods=len(df_flights), freq='M')
passageiros = df_flights["passengers"]

passageiros
```

## 7) Modelos GARCH

Veremos agora nossa última classe de modelos: **Generalized AutoRegressive Conditional Heteroskedasticity**.

O objetivo deste modelo é **modelar volatilidades variáveis de séries temporais**.

**Heterocedasticidade:**

- diferente (hetero) + dispersão (skedasis);
- series temporais demonstram variações de volatilidade sistematicamente ao longo to tempo.

A heterocedasticidade é facilmente detectável ao olharmos para a série temporal:

<img src="https://media.geeksforgeeks.org/wp-content/uploads/20190425172205/hetero.jpg" width=500>


O modelo **ARCH** busca modelar variância média ponderada dos **resíduos** passados:

$$ARCH(p): \sigma_t ^2 = \omega + \sum_{i=1}^{p}\alpha_i \epsilon_{t-1}^2$$
$$ARCH(1): \sigma_t ^2 = \omega + \alpha \epsilon_{t-1}^2$$

$$\epsilon_t = w_t\sqrt{\omega + \alpha \epsilon_{t-1}^2}$$

Já o modelo **GARCH** generaliza o ARCH ao introduzir uma componente que faça uso das próprias volatilidades passadas:

$$GARCH(p, q): \sigma_t ^2 = \omega + \sum_{i=1}^{p}\alpha_i \epsilon_{t-1}^2 + \sum_{j=1}^{q}\beta_j \sigma_{t-j}^2$$
$$GARCH(1, 1): \sigma_t ^2 = \omega + \alpha \epsilon_{t-1}^2 + \beta \sigma_{t-1}^2$$

$$\epsilon_t = w_t\sqrt{\omega + \alpha \epsilon_{t-1}^2 + \beta \sigma_{t-1}^2}$$

**Intuição**

- autoregressivo: prediz comportamento futuro baseado no comportamento passado;
- volatilidade é a média ponderada de informações passadas.

Aqui, utilizaremos o GARCH de uma forma bem aplicada: para calcular um $\beta$ dinâmico de um ativo!

Lembrando que o beta é dado por:

$$\beta_P = \frac{\Sigma(R_P, R_M)}{\sigma^2(R_M)}$$

Poder calcular as covariâncias de forma dinâmica faz com que o $\beta$ possa ser estudado de forma dinâmica no tempo, e não como um único valor que leva em consideração a série completa em todo o período!

```python
precos_ativo = DataReader('ABEV3.SA', data_source='yahoo', start='2010-01-01')['Adj Close']
precos_mercado = DataReader('^BVSP', data_source='yahoo', start='2010-01-01')['Adj Close']

retornos_ativo = precos_ativo.pct_change().dropna()
retornos_mercado = precos_mercado.pct_change().dropna()

selic_anual = 0.02
r_rf = (1 + selic_anual) ** (1/252) - 1

retornos_ex_ativo = retornos_ativo - r_rf
retornos_ex_mercado = retornos_mercado - r_rf

df_capm = pd.DataFrame({"ER_ativo": retornos_ex_ativo,
                        "ER_mercado": retornos_ex_mercado}).dropna()
```

```python
df_capm
```

```python
df_capm.cov().iloc[0, 1]/df_capm["ER_mercado"].var()
```

```python
# !pip install arch
```

```python
# importando a biblioteca

from arch import arch_model
```

```python
# fitando os modelos GARCH

gm_ativo = arch_model(df_capm["ER_ativo"], vol='Garch', p=1, o=0, q=1).fit()
gm_mercado = arch_model(df_capm["ER_mercado"], vol='Garch', p=1, o=0, q=1).fit()
```

```python
print(gm_mercado.summary())
```

```python
# pegando as volatilidades modeladas

gm_vol_ativo = gm_ativo.conditional_volatility
gm_vol_mercado = gm_mercado.conditional_volatility
```

```python
# calculando os resíduos (normalizados pela volatilidade modelada)

resid_abev = gm_ativo.resid / gm_vol_ativo
resid_ibvp = gm_mercado.resid / gm_vol_mercado
```

O beta dinâmico será dado pelo produto entre a correlação dos resíduos e a razão entre a volatilidade modelada do ativo e do mercado:

```python
np.corrcoef(resid_abev, resid_ibvp)
```

```python
pd.DataFrame({"P":resid_abev, "M":resid_ibvp}).corr()
```

```python
correlation * (gm_vol_ativo/gm_vol_mercado)
```

```python
correlation = np.corrcoef(resid_abev, resid_ibvp)[0, 1]

beta = correlation * (gm_vol_ativo/gm_vol_mercado)

beta.plot(figsize=(10, 8))
```

```python
beta.mean()
```

Lembrando que:


- $\beta < 0$: relação inversa entre ativo e mercado, o que é bem raro de ser encontrado.

- $\beta = 0$: o ativo é completamente descorrelacionado com o mercado.

- $0 < \beta < 1$: ativo menos volátil que o mercado, mas correlacionado com ele.

- $\beta = 1$: o ativo acompanha o mercado perfeitamente.

- $\beta > 1$: ativo mais volátil que o mercado.
