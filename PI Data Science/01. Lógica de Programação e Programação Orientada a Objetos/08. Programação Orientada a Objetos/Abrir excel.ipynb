```python
def trata_escol(x):
    if type(x) == str:
        if x == '.':
            return np.nan
        else:
            return x.title()
    else:
        return x
```

```python
quest['Escolaridade'] = quest['Escolaridade'].apply(trata_escol)
```

```python
for povo in quest ['Cód. Povo'].dropna().unique():
    for escolaridade in quest['Escolaridade'].dropna().unique():
        mascara = (quest['Escolaridade']==escolaridade)&(quest['Cód. Povo']==povo)
        numero = len(quest[mascara])
        contagem_escolaridade.loc[escolaridade,povo]= numero
```
