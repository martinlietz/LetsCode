# Figure
A `figure` - em portugues figura ou melhor imagem e onde fica tudo, os eixos, alguns `artes` especias como titulo, legenda, etc... e o canvas, imagina o canvas como uma lousa que vamos usar para desenhar. 
Um `figure`( sei que é dificil mas aceita que é a imagem ;-) ) pode ter varias eixos, mas por padrão tem um eixo.

A forma mais facil de criar um `figure` é `pyplot`:

```python
fig = plt.figure()  # figure vazia sem eixos
figsizePX=(600*px, 400*px) # com pixels
figsize=(10*cm, 10*cm) # com centimetros
fig, ax = plt.subplots(figsize=figsize)  # figure com um eixo
fig, axs = plt.subplots(2, 2, figsize=figsizePX)  # figure com 2x2 grid 
```

> Normalmente criamos os eixos com o figure mas voce pode criar também depois sem problemas.

Temos dus formas de criar o grafico  
1. criar explicitamente com figure, axes etc..  


```python
x = np.linspace(0, 2, 100)

# Note that even in the OO-style, we use `.pyplot.figure` to create the figure.
fig, ax = plt.subplots()  # Create a figure and an axes.
ax.plot(x, x, label='linear')  # Plot some data on the axes.
ax.plot(x, x**2, label='quadratic')  # Plot more data on the axes...
ax.plot(x, x**3, label='cubic')  # ... and some more.
ax.set_xlabel('x label')  # Add an x-label to the axes.
ax.set_ylabel('y label')  # Add a y-label to the axes.
ax.set_title("Simple Plot")  # Add a title to the axes.
ax.legend(loc=4)  # Add a legend.
```

2. confiar na lib e criar com pyplot que `advinha` que voce quer

```python
x = np.linspace(0, 2, 100)

plt.plot(x, x, label='linear')  # Plot some data on the (implicit) axes.
plt.plot(x, x**2, label='quadratic')  # etc.
plt.plot(x, x**3, label='cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend()
```

Com o tempo a gente sempre recriamos os mesmos graficos, portanto é uma boa pratica de criar uma lib proprio com os nossos estilos.  
Para isso pode definir um arquivo .py e incluindo funções como :

```python
def my_plotter(ax, data1, data2, param_dict):
    """
    A helper function to make a graph

    Parameters
    ----------
    ax : Axes
        The axes to draw to

    data1 : array
       The x data

    data2 : array
       The y data

    param_dict : dict
       Dictionary of kwargs to pass to ax.plot

    Returns
    -------
    out : list
        list of artists added
    """
    out = ax.plot(data1, data2, **param_dict)
    return out
```

assim podemos usar dentro do nosso notbook :

```python
data1, data2, data3, data4 = np.random.randn(4, 100)
fig, ax = plt.subplots(1, 1)
my_plotter(ax, data1, data2, {'marker': 'x'})
```

agora ficou facil criar duas :

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
my_plotter(ax1, data1, data2, {'marker': 'x'})
my_plotter(ax2, data3, data4, {'marker': 'o'})
```
