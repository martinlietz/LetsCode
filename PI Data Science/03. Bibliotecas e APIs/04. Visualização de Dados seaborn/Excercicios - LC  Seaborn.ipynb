```python
df_avocado.info()
```

```python
sns.displot(df_avocado[df_avocado['type']=='conventional']['AveragePrice'], kde=False)
```

```python
sns.displot(df_avocado[df_avocado['type']=='organic']['AveragePrice'], kde=False)
```

```python
sns.lmplot(data=df_avocado, x='AveragePrice', y='Total Volume', hue='type')
```

```python
corr = df_avocado.corr()
sns.heatmap(data=corr, cmap='coolwarm', annot=True)
```
