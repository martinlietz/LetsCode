 - AirTime: tempo de vôo (em minutos);
 - ArrDelay: atraso na chegada (em minutos);
 - DepDelay: atraso na saída (em mínutos);
 - Origin: Aeroporto de origem [(código IATA](https://pt.wikipedia.org/wiki/C%C3%B3digo_aeroportu%C3%A1rio_IATA));

 - Distance: em milhas;
 - TaxiIn: taxi in time, in minutes;
 - TaxiOut: taxi out time in minutes;
 - Cancelled: O vôo foi cancelado?
 - CancellationCode: Razão do cancelamento (A = empresa, B = clima, C = NAS (National Airspace System), D = segurança);
 - Diverted: 1 = sim, 0 = não;
 - CarrierDelay: em minutos;
