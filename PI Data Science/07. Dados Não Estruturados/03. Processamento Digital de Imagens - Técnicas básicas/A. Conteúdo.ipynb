<h3 id="1-técnicas-básicas-de-pdi"><strong>1) Técnicas básicas de PDI</strong></h3>
<h3 id="--tipos-de-operações">- Tipos de Operações</h3>
<p>As principais operações de PDI podem ser categorizadas em:</p>
<ul>
<li>Pixel-a-pixel;</li>
<li>De vizinhança;</li>
<li>Outras.</li>
</ul>
<p>Vamos, a seguir, exemplificar cada uma delas:</p>
<h3 id="--operações-pixel-a-pixel">- Operações pixel-a-pixel</h3>
<p>São operações realizadas em cada pixel individualmente.</p>
<h4 id="limiarização"><em>Limiarização</em></h4>
<p>Este tipo de operação é utilizado para separar níveis de cinza de uma imagem, de modo que grupos de pixels de intensidades similares sejam agrupados. Este processo ressalta detalhes da imagem na esala de cinzas, como podemos ver a seguir:</p>
<p><img src="https://i.imgur.com/FswVKL0.jpg" width="500"></p>
<h4 id="ajuste-de-contraste-e-brilho"><em>Ajuste de contraste e brilho</em></h4>
<p>O brilho de uma imagem refere-se à intensidade dos pixels. Quanto maior o btrilho, mais iluminada a imagem aparenta. Para um brilho baixo, a imagem aparenta ser opaca.</p>
<p>Já o contraste, controla a intensidade relativa entre cores claras e escuras. Quanto maior o contraste, maior é a diferença de intensidade entre cores claras e escuras, de forma que uma se sobressai com relação à outra.</p>
<p>Para melhoras a qualidade de uma imagem, é importante ajustar ambos o brilho e o contraste simultaneamente, até que a combinação ideal seja encontrada:</p>
<p><img src="https://i.imgur.com/3D7l8g4.jpg" width="500"></p>
<h3 id="--operações-de-vizinhança">- Operações de vizinhança</h3>
<p>São operações que atuam não em pixels isoladamente, mas de forma unificada em pixels vizinhos, processando um grupo de pixels simultaneamente.</p>
<h4 id="convolução"><em>Convolução</em></h4>
<p>A operação de convolução é definida como a aplicação de um filtro a determinado número de pixels, que retorna um único número, resultado de uma operação feita com os pixels aos quais o filtro foi aplicado.</p>
<p>Um exemplo de convolução consiste em somar os valores dos pixels, como exemplificado abaixo. Outro exemplo seria tomar a média dos pixels. </p>
<p>Dependendo do filtro, o resultado da convolução será diferente. Por exemplo:</p>
<ul>
<li>O filtro da média é um filtro de suavização, que torna a imagem mais uniforme, eliminando ruídos.</li>
<li>O filtro laplaciano é um filtro de detecção de bordas, que retorna uma borda fina ao redor das formas nas imagens.</li>
</ul>
<p>Como podemos ver nos exemplos a seguir, a operação de convolução garante que os filtros sejam aplicados à imagem como um todo, mas agindo localmente:</p>
<p><img src="https://i.imgur.com/DDR4Oto.jpg" width="500">
<br>
<img src="https://i.imgur.com/32gT5fW.jpg" width="500"></p>
<p>Alguns outros filtros comumente utilizado em convoluções são:</p>
<p><em>Filtro mediana</em></p>
<p>O filtro mediana também é um filtro de suavização (utilizado para remoção de ruído), e sua aplicação consiste nas seguintes etapas:</p>
<ul>
<li>Ler todos os pixels da vizinhança;</li>
<li>Ordenar os pixels em ordem crescente;</li>
<li>Retornar o pixel na posição média do ordenamento (mediana entre os pixels).<ul>
<li>Exemplo: para um filtro 3x3 = 9 pixels, o filtro retorna o pixel que estiver na posição do meio, isto é, posição 5:
<br><br>
<img src="https://i.imgur.com/t96OqTv.jpg" width="500"></li>
</ul>
</li>
</ul>
<p>Além do filtro laplaciano apresentado acima, há outros dois filtros de deteção de bordas que são comumente utilizados: Sobel e Prewitt. Estes filtros são bem mais resistentes à ruídos que o filtro Laplaciano, então, eles podem ser aplicados sem a necessidade da aplicação prévia de um suavizador.</p>
<p><em>Sobel</em></p>
<p>Este filtro é capaz de simultaneamente proporcionar deteção de bordas e suavização (redução de ruídos).</p>
<p><img src="https://i.imgur.com/pN5mU4X.jpg" width="500"></p>
<p><em>Prewitt</em></p>
<p>Diferentemente do filtro Sobel, o Prewitt não enfatiza pixels próximo ao centro do filtro, assim, não proporciona suavização da imagem.</p>
<p><img src="https://i.imgur.com/CxX3tKV.jpg" width="500"></p>
<p>Os filtros de convolução podem ser aplicados a vizinhanças de diferentes tamanhos, basta mudar as dimensões do filtro:</p>
<p><img src="https://i.imgur.com/FadVeaI.jpg" width="500"></p>
<p>Dependendo da conectividade entre os pixels, os filtros de convolução podem agir de maneiras diferentes. Temos:</p>
<ul>
<li>Pixels 4-conectados: são pixels vizinhos a pixels que tocam suas bordas, sendo conectados horizontal e verticalmente;</li>
<li>Pixels 8-conectados: são pixels vizinhos a pixels que tocam suas bordas <strong>e</strong> cantos, sendo conectados também diagonalmente.
<br><br>
<img src="https://i.imgur.com/dxZnRzD.jpg" width="500"></li>
</ul>
<p>As ações simultãneas de dois filtros podem ser alteradas de acordo com operações lógicas, que alteram a região de ação dos filtros, conforme ilustrado a seguir: </p>

<p>Por fim, há filtros que atuam ao transformar determinados atributos geométricos da imagem, como alterar seu tamanho, rotacioná-la, etc.</p>
<p>Estes filtros representam <em>transformações geométricas</em>, que são todas transformações afins, pois são representadas por matrizes afins.</p>

<p><strong>Filtros Prewitt e Sobel</strong></p>
<p>Agora, aplicamos os outros dois filtros de detecção de bordas que discutimos:</p>
