```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

**1 - Filtro da m√©dia**

dst	=	cv2.blur(src, ksize, )

src: imagem input (numpy array)

dst: imagem output (numpy array)

Ksize: tamanho do kernel: (3,3), (5,5), etc.




```python
img = cv2.imread('text.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
# aplicando o filtro com kernel (3,3)
img_filtro_media_3_3 = cv2.blur( img, ksize=(3,3) )
```

```python
plt.figure(figsize=(10,10))
plt.subplot(121)
plt.imshow(img)
plt.title('original')

plt.subplot(122)
plt.imshow(img_filtro_media_3_3)
plt.title('fitro media 3x3')
```

```python
# aplicando o filtro da media com kernel (15,15)
img_filtro_media_15_15 = cv2.blur( img, ksize=(15,15) )

plt.figure(figsize=(10,10))
plt.subplot(121)
plt.imshow(img)
plt.title('original')

plt.subplot(122)
plt.imshow(img_filtro_media_15_15)
plt.title('fitro media 15x15')
```

# Filtro gaussiano


**1 - Filtro Gaussiano**

dst	=	cv2.GaussianBlur(src, ksize, sigmaX, sigmaY)

src: imagem input (numpy array)

dst: imagem output (numpy array)

Ksize: tamanho do kernel gaussiano (tuple): (3,3), (5,5), etc.

sigmaX: Desvio padr√£o em X

sigmaY: Desvio padr√£o em Y



```python
plt.imshow(img)
```

```python
img_filtro_gaussiano_3x3 = cv2.GaussianBlur(img, ksize=(3,3), sigmaX=1, sigmaY=1)

plt.figure(figsize=(10,10))
plt.subplot(121)
plt.imshow(img)
plt.title('original')

plt.subplot(122)
plt.imshow(img_filtro_gaussiano_3x3)
plt.title('fitro gaussiano 3x3')
```

## Opera√ß√µes morfrl√≥gicas - Dilata√ß√£o

    

```python
img = cv2.imread('fechamento.png')
plt.imshow(img)
```

```python
# kernel que utilizaremos para dilata√ß√£o e eros√£o
kernel = np.ones((3,3))
kernel
```

```python
img_dilatada = cv2.dilate(img, kernel = kernel, iterations=2)

# plt.figure(figsize=(10,10))
plt.subplot(121)
plt.imshow(img)
plt.title('original')

plt.subplot(122)
plt.imshow(img_dilatada)
plt.title('img_dilatada 5x5')
```

```python
## opera√ß√£o morfol√≥gica de Eros√£o
```

```python
img = cv2.imread('jota.png')
plt.imshow(img)
```

```python
img_erodido = cv2.erode(img, kernel=kernel, iterations=2)

plt.subplot(121)
plt.imshow(img)
plt.title('original')

plt.subplot(122)
plt.imshow(img_erodido)
plt.title('img_erodido 3x3')
```

```python
img_erodido_dilatado = cv2.dilate(img_erodido, kernel=kernel, iterations=2)

plt.subplot(121)
plt.imshow(img)
plt.title('original')

plt.subplot(122)
plt.imshow(img_erodido_dilatado)
plt.title('img_erodido_dilatado 3x3')
```
