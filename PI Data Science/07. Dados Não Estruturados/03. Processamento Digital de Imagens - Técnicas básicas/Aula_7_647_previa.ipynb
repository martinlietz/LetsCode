```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('../img/smarties.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(img, cmap='gray')
```

```python
# kernel de prewitt x e y
prewitt_x = np.array([
    [-1, 0, 1],
    [-1, 0, 1],
    [-1, 0, 1],
])
prewitt_y = np.array([
    [1, 1, 1],
    [0, 0, 0],
    [-1, -1, -1],
])
```

```python
# aplicar o filtro de prewitt na direcao x
img_contorno_x = cv2.filter2D(img, ddepth=cv2.CV_32F, kernel=prewitt_x)
```

```python
# plt.figure(figsize=(10,10))
plt.subplot(121)
plt.imshow(img, cmap='gray')
plt.title('original')

plt.subplot(122)
plt.imshow(img_contorno_x, cmap='gray')
plt.title('img_contorno_x prewitt')
```

```python
# aplicar o filtro de prewitt na direcao x
img_contorno_y = cv2.filter2D(img, ddepth=cv2.CV_32F, kernel=prewitt_y)
```

```python
plt.figure(figsize=(10,10))
plt.subplot(131)
plt.imshow(img, cmap='gray')
plt.title('original')

plt.subplot(132)
plt.imshow(img_contorno_y, cmap='gray')
plt.title('img_contorno_y prewitt')
plt.subplot(133)
plt.imshow(img_contorno_x, cmap='gray')
plt.title('img_contorno_x prewitt')
```

```python
img = cv2.imread('../img/220px-Camera_obscura.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(img, cmap='gray')
```

```python
# kernel de sobel x e y
sobel_x = np.array([
    [1, 0, -1],
    [2, 0, -2],
    [1, 0, -1],
])
sobel_y = np.array([
    [1, 2, 1],
    [0, 0, 0],
    [-1, -2, -1],
])
```

```python
# aplicar o filtro de sobel na direcao x
img_cantos_x = cv2.filter2D(img, ddepth=cv2.CV_32F, kernel=sobel_x)
img_cantos_y = cv2.filter2D(img, ddepth=cv2.CV_32F, kernel=sobel_y)
img_cantos_xy = cv2.filter2D(img, ddepth=cv2.CV_32F, kernel=np.matmul(sobel_x,sobel_y))
```

```python
plt.figure(figsize=(10,10))
plt.subplot(141)
plt.imshow(img, cmap='gray')
plt.title('original')

plt.subplot(142)
plt.imshow(img_cantos_y, cmap='gray')
plt.title('img_cantos_y sobel')
plt.subplot(143)
plt.imshow(img_cantos_x, cmap='gray')
plt.title('img_cantos_x sobel')
plt.subplot(144)
plt.imshow(img_cantos_xy, cmap='gray')
plt.title('img_cantos_xy sobel')
```

```python
# consolidando as duas imagens numa √∫nica
img_sobel = cv2.magnitude(x=img_cantos_x, y=img_cantos_y)
plt.imshow(img_sobel, cmap='gray')
```

```python
img = cv2.imread('../img/filtro_mediano.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_mediano = cv2.medianBlur(img, ksize=5)

img_mediano = cv2.cvtColor(img_mediano, cv2.COLOR_RGB2GRAY)
plt.imshow(img_mediano, cmap='gray')
```

```python
img_sobel_y = cv2.Sobel(img_mediano, ddepth=cv2.CV_32F, dx=0, dy=1, ksize=3)
plt.imshow(img_sobel_y, cmap='gray')
img_sobel_x = cv2.Sobel(img_mediano, ddepth=cv2.CV_32F, dx=1, dy=0, ksize=3)
plt.subplot(121)
plt.imshow(img_sobel_y, cmap='gray')
plt.subplot(122)
plt.imshow(img_sobel_x, cmap='gray')
```

```python
# consolidando as duas imagens numa √∫nica
img_sobel = cv2.magnitude(x=img_sobel_x, y=img_sobel_y)
plt.imshow(img_sobel, cmap='gray')
```

```python
img_canny = cv2.Canny(img_sobel, threshold1=10, threshold2=100)
plt.imshow(img_canny, cmap='gray')
```
