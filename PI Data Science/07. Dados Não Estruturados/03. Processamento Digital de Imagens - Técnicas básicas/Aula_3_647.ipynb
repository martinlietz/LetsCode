```python
# desafio: aplicar o cv2.threshold escolhendo um Limiar talque consiga
# ressaltar as linhas e nÃºmeros da imagem quando binarizada.
img = cv2.imread('../img/sudoku.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
print(img.shape)
plt.imshow(img, cmap='gray')
```

```python
plt.figure(figsize=(20,30))
plt.subplot(141)
limiar = 150
retval, dst = cv2.threshold(src = img, thresh = limiar, maxval = 255, type = cv2.THRESH_BINARY )
plt.imshow(dst, cmap='gray')
plt.title(f'Limiar={limiar}')

plt.subplot(142)
limiar = 100
retval, dst = cv2.threshold(src = img, thresh = limiar, maxval = 255, type = cv2.THRESH_BINARY )
plt.imshow(dst, cmap='gray')
plt.title(f'Limiar={limiar}')

plt.subplot(143)
limiar = 50
retval, dst = cv2.threshold(src = img, thresh = limiar, maxval = 255, type = cv2.THRESH_BINARY )
plt.imshow(dst, cmap='gray')
plt.title(f'Limiar={limiar}')

plt.subplot(144)
limiar = 25
retval, dst = cv2.threshold(src = img, thresh = limiar, maxval = 255, type = cv2.THRESH_BINARY )
plt.imshow(dst, cmap='gray')
plt.title(f'Limiar={limiar}')
```

```python
# agora vamos usar o metodo de threshold adaptativo
plt.figure(figsize=(20,30))
plt.subplot(131)
blockSize = 3
dst = cv2.adaptiveThreshold( img, 
                      maxValue = 255, 
                      adaptiveMethod = cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                      thresholdType = cv2.THRESH_BINARY,
                      blockSize = blockSize,
                      C = 1)

plt.imshow(dst, cmap='gray')
plt.title(f'blockSize={blockSize}')

plt.subplot(132)
blockSize = 5
dst = cv2.adaptiveThreshold( img, 
                      maxValue = 255, 
                      adaptiveMethod = cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                      thresholdType = cv2.THRESH_BINARY,
                      blockSize = blockSize,
                      C = 1)

plt.imshow(dst, cmap='gray')
plt.title(f'blockSize={blockSize}')

plt.subplot(133)
blockSize = 7
dst = cv2.adaptiveThreshold( img, 
                      maxValue = 255, 
                      adaptiveMethod = cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                      thresholdType = cv2.THRESH_BINARY,
                      blockSize = blockSize,
                      C = 1)

plt.imshow(dst, cmap='gray')
plt.title(f'blockSize={blockSize}')
```
