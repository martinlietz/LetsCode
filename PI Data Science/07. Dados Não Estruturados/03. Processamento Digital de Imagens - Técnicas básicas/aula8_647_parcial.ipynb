```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('smarties.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img, cmap='gray')
```

```python
# aplicando o filtro de sobel no eixo x
img_sobel_x = cv2.Sobel(img, ddepth=cv2.CV_32F, dx=1, dy=0, ksize=3)
plt.imshow(img_sobel_x, cmap='gray')
```

```python
img_sobel_y = cv2.Sobel(img, ddepth=cv2.CV_32F, dx=0, dy=1, ksize=3)
plt.imshow(img_sobel_y, cmap='gray')
```

```python
# consolidando as duas imagens numa única
img_sobel = cv2.magnitude(x=img_sobel_x, y=img_sobel_y)
plt.imshow(img_sobel, cmap='gray')
```

```python
img_sobel = cv2.Sobel(img, ddepth=cv2.CV_32F, dx=1, dy=1, ksize=3)
plt.imshow(img_sobel, cmap='gray')
```

```python
# # com a imagem 'filtro_mediano.png'
# 1. aplicar o filtro de sobel na direção x
# 2. aplicar o filtro de sobel na direção y
# 3. calcular a magnetude
```

```python
img = cv2.imread('filtro_mediano.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img, cmap='gray')
```

```python
img_sobel_x = cv2.Sobel(img, ddepth=cv2.CV_32F, dx=1, dy=0, ksize=3)
img_sobel_y = cv2.Sobel(img, ddepth=cv2.CV_32F, dx=0, dy=1, ksize=3)
img_sobel = cv2.magnitude(img_sobel_x, img_sobel_y)
plt.imshow(img_sobel, cmap='gray')
plt.title('Não deu certo!')
```

```python
# método de suavização

img_mediano = cv2.medianBlur(img, ksize = 3)

img_gaussiano = cv2.GaussianBlur(img, ksize=(3,3), sigmaX=1)

img_media = cv2.filter2D(img, ddepth=cv2.CV_8U, kernel = 1/9*np.ones((3,3)))

plt.imshow(img, cmap='gray')
plt.title('original')
plt.show()

plt.figure(figsize=(20,10))
plt.subplot(131)
plt.imshow(img_mediano, cmap='gray')
plt.title('Mediano')

plt.subplot(132)
plt.imshow(img_gaussiano, cmap='gray')
plt.title('Gaussiano')

plt.subplot(133)
plt.imshow(img_media, cmap='gray')
plt.title('Media')
```

```python
# vimos que para ruidos aleatórios o filtro mediano tem um melhor resultado
# mas agora, vamos alterar o tamanho do kernel
img_mediano = cv2.medianBlur(img, ksize = 9)
plt.imshow(img, cmap='gray')
plt.title('original')
plt.show()

plt.imshow(img_mediano, cmap='gray')
plt.title('img_mediano')
plt.show()
```

```python
img_sobel_x = cv2.Sobel(img_mediano, ddepth=cv2.CV_32F, dx=1, dy=0, ksize=3)
img_sobel_y = cv2.Sobel(img_mediano, ddepth=cv2.CV_32F, dx=0, dy=1, ksize=3)
img_sobel_com_filtro_mediano = cv2.magnitude(img_sobel_x, img_sobel_y)
plt.imshow(img_sobel_com_filtro_mediano, cmap='gray')
```

```python
# a imagem fica melhor se usarmos ela colorida
img = cv2.imread('filtro_mediano.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
img_filtrada = cv2.medianBlur(img, ksize=5)
plt.imshow(img_filtrada)
```

```python
img_gray = cv2.cvtColor(img_filtrada, cv2.COLOR_RGB2GRAY)
img_sobel_x = cv2.Sobel(img_gray, ddepth=cv2.CV_32F, dx=1, dy=0, ksize=3)
img_sobel_y = cv2.Sobel(img_gray, ddepth=cv2.CV_32F, dx=0, dy=1, ksize=3)
img_sobel_com_filtro_mediano_colorido = cv2.magnitude(img_sobel_x, img_sobel_y)
plt.imshow(img_sobel_com_filtro_mediano_colorido, cmap='gray')
```

```python
plt.imshow(img_sobel, cmap='gray')
plt.show()

plt.imshow(img_sobel_com_filtro_mediano, cmap='gray')
plt.show()

plt.imshow(img_sobel_com_filtro_mediano_colorido, cmap='gray')
plt.show()
```

```python
# vamos utilizar um método estado da arte para filtrar contorno - cv2.Canny

plt.imshow(img_gray, cmap='gray')
```

```python

img_canny = cv2.Canny(img_gray, threshold1=10, threshold2=80)
plt.imshow(img_canny, cmap='gray')
plt.title('threshold1=10')
plt.show()

img_canny = cv2.Canny(img_gray, threshold1=50, threshold2=80)
plt.imshow(img_canny, cmap='gray')
plt.title('threshold1=50')
plt.show()
```

```python

img_canny = cv2.Canny(img_gray, threshold1=10, threshold2=80)
plt.imshow(img_canny, cmap='gray')
plt.title('threshold2=80')
plt.show()

img_canny = cv2.Canny(img_gray, threshold1=10, threshold2=150)
plt.imshow(img_canny, cmap='gray')
plt.title('threshold2=150')
plt.show()
```
