```python
cx = 1/4
cy = 1/4
M = np.array([
    [cx*1.0, 0, 0]
    ,[0, cy*1.0, 0]
    
    
])
img_col = cv2.imread('..\img/lena_color.png')
# como converter de BGR para RGB
img_col = cv2.cvtColor(img_col, code=cv2.COLOR_BGR2RGB)
img_escalada = cv2.warpAffine(img_col, M, dsize=(int(img_col.shape[0]*cx),int(img_col.shape[1]*cy)))
plt.imshow(img_escalada)
```

```python
# translando a imagem
tx = -20
ty = 40
M_translacao = np.array([
    [1.0, 0, tx]
    ,[0, 1.0, ty]
    
    
])
img_col = cv2.imread('..\img/lena_color.png')
# como converter de BGR para RGB
img_col = cv2.cvtColor(img_col, code=cv2.COLOR_BGR2RGB)
img_transladada = cv2.warpAffine(img_col, M_translacao, dsize=(int(img_col.shape[0]),int(img_col.shape[1])))
plt.imshow(img_transladada)
```

```python
# rotação da imagem
theta = 35
theta_rad = theta * np.pi/180
M_rotacao = np.array([
    [np.cos(theta_rad), np.sin(theta_rad), 0]
    ,[-np.sin(theta_rad), np.cos(theta_rad), 0]
    
    
])
img_col = cv2.imread('..\img/lena_color.png')
# como converter de BGR para RGB
img_col = cv2.cvtColor(img_col, code=cv2.COLOR_BGR2RGB)
img_rotacionado = cv2.warpAffine(img_col, M_rotacao, dsize=(int(img_col.shape[0]),int(img_col.shape[1])))
plt.imshow(img_rotacionado)
```

```python
# somando com um escalar beta = 1
img_col = cv2.imread('..\img/lena_color.png')
# como converter de BGR para RGB
img_col = cv2.cvtColor(img_col, code=cv2.COLOR_BGR2RGB)
beta = 50

f = img_col 
g = img_col + beta
plt.subplot(121),plt.imshow(f)
plt.subplot(122),plt.imshow(g)
plt.title(f'somando com um escalar beta {beta}'  )
```

```python
# somando com um escalar beta = 1
img_col = cv2.imread('..\img/lena_color.png')
# como converter de BGR para RGB
img_col = cv2.cvtColor(img_col, code=cv2.COLOR_BGR2RGB)
beta = 200
beta_matriz = np.ones(img_col.shape,dtype='uint8')*beta
f = img_col 
g = cv2.add(img_col, beta)
plt.subplot(121)
plt.imshow(f)
plt.subplot(122)
plt.imshow(g)
plt.title(f'beta {beta}'  )
```

```python
# somando com um escalar beta = 1
img_col = cv2.imread('..\img/lena_color.png')
# como converter de BGR para RGB
img_col = cv2.cvtColor(img_col, code=cv2.COLOR_BGR2RGB)
beta = 200
beta_matriz = np.ones(img_col.shape,dtype='uint8')*beta
f = img_col 
g = cv2.subtract(img_col, beta)
plt.subplot(121)
plt.imshow(f)
plt.subplot(122)
plt.imshow(g)
plt.title(f'beta {beta}'  )
```

```python
# alterando o cotraste> multiplicar por um escalar alpha
img_col = cv2.imread('..\img/lena_color.png')
# como converter de BGR para RGB
img_col = cv2.cvtColor(img_col, code=cv2.COLOR_BGR2RGB)
#beta = 200
#beta_matriz = np.ones(img_col.shape,dtype='uint8')*beta
alpha = 2
alpha_matriz = np.ones(img_col.shape,dtype='uint8')*alpha
f = img_col 
g = cv2.multiply(img_col, alpha_matriz)
plt.subplot(121)
plt.imshow(f)
plt.subplot(122)
plt.imshow(g)
plt.title(f'alpha {alpha}'  )
```

```python
# alterando o cotraste> multiplicar por um escalar alpha
img_col = cv2.imread('..\img/lena_color.png')
# como converter de BGR para RGB
img_col = cv2.cvtColor(img_col, code=cv2.COLOR_BGR2RGB)
h = np.zeros((img_col.shape), dtype='uint8')
mascara = cv2.rectangle(img=h, pt1=(50,75), pt2=(150,150), color=(255,255,255), thickness=-1)
plt.subplot(121)
plt.imshow(mascara)
plt.subplot(122)
plt.imshow(img_col)
plt.title(f'mascara'  )
```

```python
f = img_col
g = cv2.bitwise_and(f,h,mascara)
plt.subplot(121)
plt.imshow(g)
plt.subplot(122)
plt.imshow(img_col)
plt.title(f'g')
```
