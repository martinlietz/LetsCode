<h3 id="1-o-que-é-pds"><strong>1) O que é PDS</strong></h3>
<p>O Processamento Digital de Sinais (PDS) é a área de estudo que trabalha com processamento de sinais, podendo eles ser: som, imagens, textos entre outros.</p>
<p>Para que os sinais possam ser analisados e processados, são usadas representações matemáticas destes, de modo que operações algébricas possam ser aplicadas em seu processamento, possibilitando, assim, a extração de informações.</p>
<p>Neste material, vamos focar principalmente bno processamento de <strong>sinais de áudio</strong>. O processamento digital de imagens (PDI) e de linguagem natural (PLN) serão discutidos em materiais próprios.</p>
<h3 id="o-que-é-um-sinal-de-áudio"><strong>O que é um sinal de Áudio?</strong></h3>
<p>O <strong>som</strong> é um fenômeno ondulatório que consiste na perturbação (compressão e expansão) periódica de moléculas de ar sob a ação de alguma fonte sonora (voz, fone de ouvido, alto-falante, etc.)</p>
<p>Tal perturbação viaja pelo ar através de ondas mecânicas esféricas, as <strong>ondas sonoras</strong>.</p>
<p>No exemplo abaixo, a vibração do pistão produz uma pressão longitudinal no ar que é percebida como som pelo ouvido humano. </p>

<p>A amplitude de uma onda refere-se à intensidade da perturbação.</p>
<p>A unidade utilizada para a medida depende do tipo da onda. No caso de ondas sonoras e sinais de áudio, a unidade utilizada é o decibel (dB). A intensidade sonora é percebida como o "volume" de um som.</p>

<p>Se as ondas estão completamente fora de fase, há <strong>interferência destrutiva</strong>: subtração das amplitudes e destruição da onda.</p>

<h3 id="3-passos-fundamentais-do-pds"><strong>3) Passos fundamentais do PDS</strong></h3>
<p>Para que uma onda sonora possa ser representada no computador, é necessário que ela sofra o processo de <strong>digitalização</strong>, que consiste nos dois passos fundamentais a seguir:</p>
<ul>
<li><p>1) <strong>Amostragem/Discretização (Sampling)</strong>: divisão da onda sonora em faixas de igual duração. Normalmente, os áudios são digitalizados a taxas de frequência entre 8kHz e 48kHz. Em um CD, o áudio é digitalizado à taxa de amostragem de 44.1kHz, ou seja, o som é divido em intervalos de 1/44.100 segundos, o que equivale a dizer que há 44.100 amostras de som em um segundo!</p>
</li>
<li><p>2) <strong>Quantização (quantizing) e codificação</strong>: cosnsite em tomar a amplitude do som nos pontos em que houve discretização, e armazenar esses valores de amplitude em uma lista numérica. Na quantização uniforme, o sinal é codificado em valores 8 bits (256 níveis) ou 16 bits (65.536 níveis). </p>
</li>
</ul>
<p>Na imagem a seguir, as barras verticais representam a amostragem, que divide a onda sonora em várias faixas de tempo fixado pela taxa de amostragem.</p>
<p>Os pontos azuis representam a amplitude nos pontos de discretização, que pelo processo de quantização são armazenados no array numérico, a representação digital da onda sonora.</p>

```python
# importando as bibliotecas necessárias
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt

#definindo as características da onda sonora
fs  = 44000 # amostragem de 44kHz. Neste caso, essa taxa será usada para a reprodução do áudio
A   = 2     # Amplitude (arbitrário)
t_i = 1     # tempo do sinal em segundos
f0  = 1200  # 1.2kHz frequência fundamental do áudio, dentro do espectro audível.
t   = np.arange(0, t_i, 1.0 / fs) # quantização uniforme de fs partes em 1 segundo

# definição da onda sonora com as características acima
wav = A * np.sin(2 * np.pi * f0 * t ) # onda senoidal

# vamos plotar para ver a onda sonora
plt.plot(t, wav)
plt.xlabel("Tempo (s)")
plt.ylabel("Amplitude")
plt.axis([0, t_i, -A, A])
plt.show()
```

```python
# importando a função Audio
from IPython.display import Audio
# reproduzindo o áudio gerado
# o primeiro argumento é a onda sonora, e o segundo é a taxa de amostragem para a reprodução
Audio(wav, rate=fs)
```
