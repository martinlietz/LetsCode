<h3 id="1-objetivos-do-pdi"><strong>1) Objetivos do PDI</strong></h3>
<p>Os principais objetivos do PDI são:</p>
<ul>
<li>Facilitar a interpretação das imagens por pessoas;</li>
<li>Permitir o armazenamento e transmissão de imagens, possibilitar sua exibição e a extração de informações;</li>
<li>Grande variedade de aplicações:<ul>
<li>Astronomia, biologia, geologia, geografia, medicina, direito, defesa, inspeção industrial, instrumentação, etc.;</li>
<li>Diferentes modalidades de acordo com a fonte da imagem: visual, Raios X (tomografia computadorizada), computador (imagens sintéticas), etc.;</li>
</ul>
</li>
<li>PDI é uma subárea de <strong>processamento digital de sinais</strong>;<ul>
<li>Originalmente, os dados processados por computadores eram essencialmente numéricos, contudo, hoje em dia, é comum ter voz, música, imagens, vídeo, etc.;</li>
</ul>
</li>
</ul>
<p>O objetivo deste módulo é apresentar noções dos métodos tradicionais de PDI, para possibilitar a escolha do algoritmo que possa resolver um problema específico.</p>
<h3 id="2-história"><strong>2) História</strong></h3>
<ul>
<li>Primeiras aplicações:<ul>
<li>Indústria jornalística: transmissões de imagens via cabo submarino entre Londres e Nova Iorque (1920);</li>
<li>Sistema de Bartlane usava 5 níveis, sendo expandido para 15 níveis em 1929.</li>
</ul>
</li>
<li>Primeiros computadores poderosos o suficiente para fazer possibilitar o PDI: anos 60;</li>
<li>No final dos anos 60 e início dos anos 70, PDI começou a ser utilizado em imagens médicas, astronômicas, etc.</li>
</ul>
<h3 id="3-definições"><strong>3) Definições</strong></h3>
<ul>
<li><p>PDI consiste em: aplicar operações matemáticas para alterar os valores dos pixels de uma
ou mais imagens.</p>
</li>
<li><p>Importância:</p>
<ul>
<li>Observar uma imagem, apenas olhando os números que descrevem os pixels, é um processo que não fornece informações relevantes;</li>
<li>O processamento de imagens é necessário para ressaltar informações importantes através da modificação de várias regiões da imagem.</li>
</ul>
</li>

<h4 id="--formação-de-cores">- Formação de cores:</h4>
<ul>
<li>No ser humano, todas as cores são percebidas como combinação das cores vermelho (R), verde (G) e azul (B).</li>
<li>As cores primárias da luz são definidas como os seguintes comprimentos de onda:<ul>
<li>R: 700nm; G: 546.1nm; B: 435,8nm.</li>
</ul>
</li>
<li>As cores primárias de pigmentos são definidas como cores que absorvem uma cor primária da luz e reflete as outras duas:<ul>
<li>Magenta = R + B;</li>
<li>Cyan = G + B;</li>
<li>Yellow = R + G.</li>
</ul>
</li>
</ul>
<h4 id="--efeitos-de-quantização">- Efeitos de quantização:</h4>

<h4 id="--tipos-de-imagem">- Tipos de Imagem:</h4>
<p>Os três principais tipos de imagem são Monocromática (preto e branco, unicamente); Níveis de cinza (preto e branco com tons de cinza intermediários); Cor 24 bits (imagem colorida com os 3 níveis RGB).</p>

<p>A depender do número de pixels em uma imagem, a percebemos com diferentes graus de resolução. </p>
<p>Uma imagem com poucos píxels tem uma aparência "serrilhada", pois é possível identificar os pixels individuais. Conforme o número de píxels aumenta, a imagem aparenta ser mais nítida:</p>

<h4 id="--pré-processamento">- Pré-processamento:</h4>
<p>O principal objetivo do pré-processamento é melhorar a imagem para possibilitar maior chance de sucesso dos processos seguintes. </p>
<p>Nesta etapa, são ressaltadas informações importantes como por exemplo bordas e objetos. Além disso, ruídos são atenuados a fim de melhorar imperfeições da imagem.</p>
<p><img src="https://i.imgur.com/CYFlAeU.jpg" width="500">
<img src="https://i.imgur.com/vmNF0wV.jpg" width="500"></p>
<p>Há diversos problemas que podem ser endereçados na etapa de pré-processamento:</p>
<ul>
<li>Problema: Ruído aleatório de aquisição ou transmissão;</li>
<li>Solução: Filtros espaciais.
<br><br>
<img src="https://i.imgur.com/uAXhtuk.jpg" width="500">
<br><br></li>
<li>Problema: Imagem com pouca variedade de cores;</li>
<li>Solução: Equalização de histograma.
<br><br>
<img src="https://i.imgur.com/dr3Ale5.jpg" width="500">
<br><br></li>
<li>Problema: Imagem corrompida devido a um movimento na câmera;</li>
<li>Solução: Realizar uma restauração a partir de um filtro de Wiener.
<br><br>
<img src="https://i.imgur.com/3yuzjBI.jpg" width="500"></li>
</ul>
<h4 id="--segmentação">- Segmentação:</h4>
<p>O processo de segmentação divide a imagem de entrada em partes ou objetos constituintes. A saída do processo de segmentação é constituída tipicamente por dados em forma de pixels, correspondendo tanto à fronteira de uma região como a todos os pontos dentro dela.</p>
<p><img src="https://i.imgur.com/kaXDROt.jpg" width="500"></p>
<p>No exemplo acima, após a segmentação é possível diferenciar em destaque os diferentes elementos da imagem de entrada.</p>
<h4 id="--extração-de-atributos">- Extração de Atributos</h4>
<p>O processo de extração de atributos tem como objetivo extrair características que resultem em alguma informação quantitativa de interesse ou que sejam básicas para discriminação entre classes de objetos de uma imagem. </p>
<p>Assim, após a extração de atributos, temos uma representação matemática vetorial dos elementos da imagem (vetor de características).</p>

<p><strong>Carregar imagem e exibir imagem</strong></p>
<p>Vamos importar a famosa imagem "Lenna". Esta imagem é uma foto da modelo sueca Lena Forsén, que é tradicionalmente utilizada para testes e ilustrações na área de PDI. Importaremos a imagem colorida:</p>
<pre class=" language-python"><code class=" language-python"><span class="token comment"># aqui a imagem é lida (salve-a no mesmo diretório do código!) e armazenada na variável "imagem"</span>

<p>Com o OpenCV é bem fácil aplicar um filtro à imagem. A seguir, aplicamos o <em>filtro mediano</em>, com o fim de eliminar ruídos da imagem, uma importante etapa de pré-processamento:</p>
<pre class=" language-python"><code class=" language-python"><span class="token comment"># aqui aplicamos o filtro mediano, através da função "medianBlur"</span>
