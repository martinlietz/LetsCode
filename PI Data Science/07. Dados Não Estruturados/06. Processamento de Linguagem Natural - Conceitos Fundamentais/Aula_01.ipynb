```python
# carregar imagem
img = cv2.imread('../img/text.jpg')
# como converter de BGR para RGB
img = cv2.cvtColor(img, code=cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
img_filtro_media_3_3 = cv2.blur(img, ksize=(3,3))
img_filtro_media_5_5 = cv2.blur(img, ksize=(5,5))
img_filtro_media_9_9 = cv2.blur(img, ksize=(9,9))
img_filtro_media_15_15 = cv2.blur(img, ksize=(15,15))
plt.figure(figsize=(25,5))
plt.subplot(151)
plt.imshow(img)
plt.subplot(152)
plt.imshow(img_filtro_media_3_3)
plt.subplot(153)
plt.imshow(img_filtro_media_5_5)
plt.subplot(154)
plt.imshow(img_filtro_media_9_9)
plt.subplot(155)
plt.imshow(img_filtro_media_15_15)
```

```python
# carregar imagem
img = cv2.imread('../img/filtro_mediano.png')
# como converter de BGR para RGB
img = cv2.cvtColor(img, code=cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
img_filtro_GaussianBlur_3_3 = cv2.GaussianBlur(src=img, ksize=(3,3),sigmaX= 0, sigmaY=0, borderType=cv2.BORDER_DEFAULT)
img_filtro_GaussianBlur_5_5 = cv2.GaussianBlur(src=img, ksize=(5,5),sigmaX= 0, sigmaY=0, borderType=cv2.BORDER_DEFAULT)
img_filtro_GaussianBlur_9_9 = cv2.GaussianBlur(src=img, ksize=(9,9),sigmaX= 0, sigmaY=0, borderType=cv2.BORDER_DEFAULT)
img_filtro_GaussianBlur_25_25 = cv2.GaussianBlur(src=img, ksize=(25,25), sigmaX= 0, sigmaY=0, borderType=cv2.BORDER_DEFAULT)
plt.figure(figsize=(25,5))
plt.subplot(151)
plt.imshow(img)
plt.subplot(152)
plt.imshow(img_filtro_GaussianBlur_3_3)
plt.subplot(153)
plt.imshow(img_filtro_GaussianBlur_5_5)
plt.subplot(154)
plt.imshow(img_filtro_GaussianBlur_9_9)
plt.subplot(155)
plt.imshow(img_filtro_GaussianBlur_15_15)
```

```python
img_filtro_mediana_5x5 = cv2.medianBlur(img, ksize = 5)

plt.figure(figsize=(10,10))
plt.subplot(121)
plt.imshow(img)
plt.title('original')

plt.subplot(122)
plt.imshow(img_filtro_mediana_5x5)
plt.title('fitro mediano 5x5')
```

```python
# Operações morfológicos Dilatação
# carregar imagem
img = cv2.imread('../img/fechamento.png')
# como converter de BGR para RGB
img = cv2.cvtColor(img, code=cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
kernel = np.ones((3,3))
```

```python
img_dilatada = cv2.dilate(img, kernel=kernel, iterations=1)
plt.subplot(121)

plt.imshow(img)
plt.subplot(122)
plt.imshow(img_dilatada)
```

```python

kernel = np.ones((5,5))
img_dilatada = cv2.dilate(img, kernel=kernel, iterations=2)
plt.subplot(121)

plt.imshow(img)
plt.subplot(122)
plt.imshow(img_dilatada)
```

```python
# Operações morfológicos Erosão ( on contrario da diltatação)
# carregar imagem
img = cv2.imread('../img/jota.png')
# como converter de BGR para RGB
#img = cv2.cvtColor(img, code=cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
kernel = np.ones((2,2))
img_erodida = cv2.erode(img, kernel=kernel, iterations=1)
plt.subplot(121)
plt.title("original")
plt.imshow(img)
plt.subplot(122)
plt.title("")
plt.imshow(img_erodida)
```

```python
kernel = np.ones((5,5))
img_dilatada_apos_eroido = cv2.dilate(img_erodida, kernel=(3,3), iterations=1)
plt.subplot(121)
plt.title("original")
plt.imshow(img)
plt.subplot(122)
plt.title("dilatado apos eroido")
plt.imshow(img_dilatada_apos_eroido)
```

```python
# Operações morfológicos de abertura ( erosão seguida de dilatação)
# carregar imagem
img = cv2.imread('../img/jota.png')
# como converter de BGR para RGB
img = cv2.cvtColor(img, code=cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
kernel = np.ones((3,3))
img_morph_abertura = cv2.morphologyEx(img, op=cv2.MORPH_OPEN, kernel=kernel, iterations=2)
plt.subplot(121)
plt.title("original")
plt.imshow(img)
plt.subplot(122)
plt.title("dilatado apos eroido")
plt.imshow(img_morph_abertura)
```
