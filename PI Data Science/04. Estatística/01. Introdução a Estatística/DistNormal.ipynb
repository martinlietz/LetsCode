# O que é a função norm.rvs?
É um função utilizada para gerar números aleatórios de acordo com u

# Onde encontro?
A função norm.rvs está localizada dentro da biblioteca ```scipy.stats```.   
Essa biblioteca contém diversos recursos estatísticos / científicos.   
Geralmente para utilizar essa função utilizamos:

````python
from scipy.stats import norm
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
norm.rvs()
````

Utilizando um desvio de ±0,5 com uma amostra de 1000, obtemos o gráfico abaixo:

````python
s = pd.Series(tuple(norm.rvs(0, 0.5, size=1000)))
s.head()

x = np.linspace(-5,5,1000)
fig, ax1 = plt.subplots(figsize=(8,6))
sns.kdeplot(s, ax=ax1)
ax1.set_xlim((-2, 2))
ax2 = ax1.twinx()
sns.histplot(data=s, ax=ax2, label='test')
plt.axvline(x=s.std()*3, c='red', linestyle='dashed')
ax1.text(s.std()*3, -.05, '6σ', color='red', transform=ax1.get_xaxis_transform(),
            ha='center', va='top')
ax1.text(s.std()*-3, -.05, '6σ', color='red', transform=ax1.get_xaxis_transform(),
            ha='center', va='top')
plt.axvline(x=s.std()*-3, c='red', linestyle='dashed')
plt.show()
````

Perfeito! Mas e se a tolerância fosse assimétrica?
Por exemplo -0,2/+0,8mm.
Neste caso o nosso intervalo da distribuição mudou.  
$\Large \frac{+0,8-(-0,2)}{2}=\frac{1,0}{2}=0,5$    
Portanto o “scale” não será mais 0,8/3, será 0,5/3.  
E o “offset” será o quanto a amostra está descentralizada.
Para o nosso exemplo deverá ser 0,3, que é:
valor min + scale ou valor max -scale

````python
s.mean()
````

````python

s1=-0.2
s2=0.8
scale=(s2-s1)/2
mu=s2-scale
sigma=6
n  = norm.rvs(mu, scale/3, size=1000)
s = pd.Series(tuple(n))
x = np.linspace(-5,5,1000)
fig, ax1 = plt.subplots(figsize=(8,6))
sns.kdeplot(s, ax=ax1)
ax1.set_xlim((-0.75,1.3 ))
ax2 = ax1.twinx()
sixsigma = s.std()*sigma/2-mu
sixsigma2 = s.std()*sigma/2+mu
sns.histplot(data=s, ax=ax2, label='test')
plt.axvline(x=-sixsigma, c='red', linestyle='dashed')
ax1.text(-sixsigma, -.05, '6σ', color='red', transform=ax1.get_xaxis_transform(),
            ha='center', va='top')
ax1.text(sixsigma2, -.05, '6σ', color='red', transform=ax1.get_xaxis_transform(),
            ha='center', va='top')
plt.axvline(x=sixsigma2, c='red', linestyle='dashed')
plt.title("6σ com tôlerancia assimétrica")
plt.show()
````
