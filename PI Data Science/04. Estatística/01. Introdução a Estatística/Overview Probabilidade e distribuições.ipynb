# Python Probabilidade, distribuição Normal, binominal, poisson e Bernoulli

A distriubuição de probabilidade nos diz quanto é provavél de uma saida x a condição de uma entrada y= $f(x)$   
Por exemplo um dado:
Podemos ter duas opções,cabeça ou numero == N = 2
Cada jogada tenho um resultado = 1
Portanta a probabilidade == $\frac{1}{2} = 0.5$


## Agora vamos ver as distribuições em python:

## Distribuição Normal

```python
import scipy.stats
from scipy.stats import lognorm
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(1234)
samples = np.random.lognormal(mean=1, sigma=.4, size=10000)
shape, loc, scale = lognorm.fit(samples, floc=0)
num_bins = 50
color = "#EFEFEF"
counts, edges, patches = plt.hist(samples, bins=num_bins, color= color)
centers = 0.5*(edges[:-1]+ edges[1:])
cdf = lognorm.cdf(edges,shape, loc=loc, scale=scale)
prob = np.diff(cdf)
plt.plot(centers, samples.size*prob, 'k-', linewidth=2)
plt.show()
```

# Binomial 

```python
import seaborn as sns
from scipy.stats import binom
data = binom.rvs(n=17, p=0.7, loc=0, size=1010)
ax = sns.displot(data, kde=True, color='pink', linewidth= 22, alpha=0.77)
ax.set(xlabel='Binomial', ylabel='Frequency')
```
