```python
# para remover os outlines, faz a distribuicao normal
describe = imoveis_sp.Valor.describe()
describe
```

```python
Q1 = describe['25%']
Q2 = describe['50%']
#Q2 = imoveis_sp.Valor.median()
Q3 = describe['75%']
```

```python
IQR = Q3 - Q1
LI = Q1 - 1.5*IQR
LS = Q3 + 1.5*IQR
```

```python
# recortar o valor maior que o LS (limite superio) e menor que o LI (limite inferior)
# a barra "|"e igual ao 'ou'
# & e a mesma coisa que ou
mascara = (imoveis_sp.Valor >= LI) & (imoveis_sp.Valor < LS)
mascara
```

```python
imoveis_sp.Valor [mascara]
```

```python
imoveis_sp.Valor [mascara].plot(kind='box')
```
