```python
# Join de duas tabelas gerando a nova book com campos unidos.
query = "SELECT B.original_title, AVG(R.rating) FROM books AS B"\
" INNER JOIN ratings as R ON B.book_id=R.book_id GROUP BY B.original_title"
book = pd.read_sql_query(sql=query, con=engine)
```
