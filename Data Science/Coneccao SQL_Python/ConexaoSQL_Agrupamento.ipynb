```python
# Trata a coluna valor ou receita de string para valor
def trata_revenue(x):
    if type(x) == str:
        x = x.replace("$","")
        x = x.replace(",","")
        x = float(x)
        return x
    elif(type(x)== float) or (type(x)== int): #Se Revenue(x) for float ou int retorna x
        return x
    else:
        return np.nan # mantem o valor 
```

```python
# Qual a venda total do setor? As vendas sÃ£o apenas de 1 setor no banco de dados.
# O apply e que vai transformar os dados em valores
sales['Revenue'] = sales.Revenue.apply(trata_revenue)
```
