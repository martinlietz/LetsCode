```python
freq_spam = {}
for palavra in spam_tratado:
    if palavra not in freq_spam.keys():
        freq_spam[palavra] = 1
    else:
        freq_spam[palavra] += 1

freq_ham = {}
for palavra in ham_tratado:
    if palavra not in freq_ham.keys():
        freq_ham[palavra] = 1
    else:
        freq_ham[palavra] += 1
```

```python
serie_ham = pd.Series(freq_ham)
serie_spam = pd.Series(freq_spam)
```

```python
mais_ocorrentes_ham = (serie_ham.sort_values(ascending=False)[:50].index)
mais_ocorrentes_spam = (serie_spam.sort_values(ascending=False)[:50].index)
```

```python
email
```

```python
def ocorrencia(x):
    ocorrencia_ = []
    if x['type'] == 'ham':
        for palavra in mais_ocorrentes_ham:
            if palavra in x['text']:
                ocorrencia_.append(palavra)
    elif x['type'] == 'spam':
        for palavra in mais_ocorrentes_spam:
            if palavra in x['text']:
                ocorrencia_.append(palavra)
    return ocorrencia_
```

```python
tabela = email.apply(ocorrencia, axis=1)
```

```python
email.head()
```

```python
tabela
```

```python
from sklearn.preprocessing import MultiLabelBinarizer
```

```python
mlb = MultiLabelBinarizer()
X = pd.DataFrame(data = mlb.fit_transform(tabela), columns=mlb.classes_, index=email.index)
```

```python
y = email.type.astype('category').cat.codes
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import BernoulliNB
from sklearn.metrics import classification_report, accuracy_score
```

```python
X_treino, X_teste, y_treino, y_teste = train_test_split(X,y)
modelo = BernoulliNB()
modelo.fit(X_treino, y_treino)
y_pred = modelo.predict(X_teste)
print(accuracy_score(y_teste, y_pred))
print(classification_report(y_teste, y_pred))
```

```python
from scipy.spatial import distance_matrix
```

```python
import seaborn as sns
iris = sns.load_dataset('iris')
```

```python
X = iris.drop('species', axis=1)
```

```python
distancias = distance_matrix(X, X)
```

```python
fig, axes = plt.subplots(ncols=2, nrows=1, figsize=(15,5))
sns.scatterplot(data=iris, x='petal_length', y='petal_width', hue='species', ax=axes[0])
sns.heatmap(distancias, ax=axes[1])
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=3)
```

```python
kmeans.fit(X)
```

```python
y= kmeans.predict(X)
```

```python
kmeans.inertia_    # qto maior pior o resultado porque é mais distante do centro
```

```python
lista_clusters = range(2,11)
inercia = []
for i in lista_clusters:
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(X)
    y= kmeans.predict(X)
    inercia.append(kmeans.inertia_)
plt.plot(lista_clusters, inercia)
```
