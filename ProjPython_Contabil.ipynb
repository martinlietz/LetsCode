Importação da bliblioteca pandas.

```python
import pandas as pd
```

Carregando o arquivo csv para um DataFrame "df_origem".

```python
df_origem = pd.read_csv("TitulosPagos_Itau.csv", delimiter=";", encoding='cp1252', header=0)
```

Criando uma coluna "ID" incrementando números sequenciais.
O software destinatário requer a sequência númerica.


```python
df_origem['ID'] = range(1, len(df_origem)+1)
```

Renomeando a coluna chave que será utilizada na funcão merge. Necessário pois a função pede um campo chave em comum. O parametro inplace força a atualização no DataFrame utilizado.

```python
df_origem.rename({"Pagador": "ORIGEM_BANCO"}, axis=1, inplace=True)
df_origem
```

Carregando o arquivo csv "De Para".

```python
df_de_para = pd.read_csv("Contabil_Layout_DePara.csv", delimiter=";", encoding='utf-8', header=0)
df_de_para
```

Realizar o merge entre os DataFrames de origem e de para.

```python
df_saida = pd.merge(df_origem, df_de_para, on='ORIGEM_BANCO', how='inner')
df_saida.head(20)
```

Removendo as colunas desnecessárias.

```python
df_saida.drop(['Vencimento','Carteira','Nosso Número', 'Seu Número', 'Situação da cobrança','Tipo de cobrança','VALOR'], axis=1, inplace=True)
```

Renomeando nome das colulas.

```python
df_saida.rename({"Data de pagamento": "DATA", "Valor(R$)": "VALOR", "ID": "LANCAMENTO", "Valor(R$)": "VALOR", }, axis=1, inplace=True)
```

Criar as colunas CCDB, CCCR e CNPJ.


```python
df_saida['CCDB'] = ''
df_saida['CCCR'] = ''
df_saida['CNPJ'] = ''
```

Reorganizando as colunas.

```python
df_saida = df_saida[['LANCAMENTO', 'DATA', 'DEBITO', 'CREDITO', 'VALOR', 'HIST_PADRAO','COMPLEMENTO', 'CCDB', 'CCCR', 'CNPJ']]
```

Ordenando por ID.

```python
df_saida = df_saida.sort_values(by='LANCAMENTO')
```

Gerando o arquivo csv final.

```python
df_saida.to_csv('Contabil.csv', index = False, sep=';',encoding='utf-8')
```
